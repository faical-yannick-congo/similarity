```python
def allclose_duplicates(df, cols, fcols, rtol=0.0, atol=0.0):
    """Determine duplicates in dataframe based on tolerances
    
    Args:
      df: the dataframe
      cols: the columns that are tested for uniquness (do not use tolerances)
      fcols: the columns that are checked for tolerance
      rtol: the relative tolerance, can be float or array with length of fcols
      atol: the absolute tolerance, can be float or array with length of fcols
    """
    def not_allclose(arr):
        return pipe(
            arr,
            np.array,
            np.absolute,
            lambda x: (x > atol + rtol * x).any()
        
        )
    
    return pipe(
        df,
        lambda x: x.duplicated(subset=cols, keep=False),
        lambda x: df.assign(duplicates=x),
        lambda x: x[x.duplicates],
        lambda x: x.sort_values(fcols),
        lambda x: x[fcols].diff().iloc[1:],
        lambda x: x.apply(not_allclose, axis=1)
    )
    
```
